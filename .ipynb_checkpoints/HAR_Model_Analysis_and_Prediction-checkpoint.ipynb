## Dataset Analysis and Pre-processing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Function to load the data
def load_data(prefix):
    X = pd.read_csv(f'{prefix}/X_{prefix}.txt', delim_whitespace=True, header=None)
    y = pd.read_csv(f'{prefix}/y_{prefix}.txt', delim_whitespace=True, header=None, names = ['Activity'])
    
    return X, y
```

```python
# Load train and test data
X_train, y_train = load_data('train')
X_test, y_test = load_data('test')
```

```python
# Load the activity labels
activity_labels = pd.read_csv('activity_labels.txt', delim_whitespace=True, header=None, names=['Activity', 'Activity_Name'])

# Merge the labels with the activity names
y_train = y_train.merge(activity_labels, how='left', on='Activity')
y_test = y_test.merge(activity_labels, how='left', on='Activity')
```

```python
# Check the shape of the data
print(f"Training data shape: {X_train.shape}")
print(f"Test data shape: {X_test.shape}")
```

```python
# Look at the head of the training data
print("Training data:")
print(X_train.head())
```

```python
# Look at the head of the training labels
print("Training labels:")
print(y_train.head())
```

```python
# Check for missing values in the training data
print("Missing values in training data:", X_train.isnull().sum().sum())
print("Missing values in training labels:", y_train.isnull().sum().sum())
```

```python
# Check for missing values in the test data
print("Missing values in test data:", X_test.isnull().sum().sum())
print("Missing values in test labels:", y_test.isnull().sum().sum())
```

```python
# Visualize the distribution of activities
plt.figure(figsize=(10, 6))
sns.countplot(y='Activity_Name', data=y_train)
plt.title("Activity distribution in training data")
plt.show()
```

## Feature Extraction

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
# Standardize the features
sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
# Apply PCA for feature extraction/dimensionality reduction
pca = PCA(n_components=0.95)  # keep 95% of variance
X_train_pca = pca.fit_transform(X_train_std)
X_test_pca = pca.transform(X_test_std)


# Let's check the shape of the datasets now
print("Shape of X_train_pca:", X_train_pca.shape)
print("Shape of X_test_pca:", X_test_pca.shape)
```

## Model Development and Evaluation

```python
# Create the Random Forest classifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
# Train the model
clf.fit(X_train_pca, y_train['Activity'])

# Use the model to predict the test labels
y_pred = clf.predict(X_test_pca)
```

```python
# Evaluate the model
print("Classification report:")
print(classification_report(y_test['Activity'], y_pred))

print("Confusion matrix:")
print(confusion_matrix(y_test['Activity'], y_pred))

print("Accuracy score:")
print(accuracy_score(y_test['Activity'], y_pred))
```

```python
# Visualize confusion matrix
plt.figure(figsize=(10, 10))
sns.heatmap(confusion_matrix(y_test['Activity'], y_pred), annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.ylabel("True Label")
plt.xlabel("Predicted Label")
plt.show()
```

### Random Forest classifier using GridSearchCV.

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Define the parameter grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 30, 50],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}
```

```python
# Create the GridSearchCV object
grid_search = GridSearchCV(estimator=clf, param_grid=param_grid, cv=5)
```

```python
# Fit the GridSearchCV object to the data
grid_search.fit(X_train_pca, y_train['Activity'])
```

```python
# Get the best parameters
print("Best parameters:", grid_search.best_params_)
```

```python
# Create the Random Forest classifier with the best parameters
clf = RandomForestClassifier(n_estimators=grid_search.best_params_['n_estimators'],
                             max_depth=grid_search.best_params_['max_depth'],
                             min_samples_split=grid_search.best_params_['min_samples_split'],
                             min_samples_leaf=grid_search.best_params_['min_samples_leaf'],
                             random_state=42)
```

```python
# Train the model
clf.fit(X_train_pca, y_train['Activity'])
```

```python
# Use the model to predict the test labels
y_pred = clf.predict(X_test_pca)
```

```python
# Evaluate the model
print("Classification report:")
print(classification_report(y_test['Activity'], y_pred))

print("Confusion matrix:")
print(confusion_matrix(y_test['Activity'], y_pred))

print("Accuracy score:")
print(accuracy_score(y_test['Activity'], y_pred))
```

## 1D Convolutional Neural Network (CNN) 

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv1D, MaxPooling1D
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
# Load the data
X_train = np.loadtxt('./train/X_train.txt')
y_train_text = np.loadtxt('./train/y_train.txt', dtype=str)

X_test = np.loadtxt('./test/X_test.txt')
y_test_text = np.loadtxt('./test/y_test.txt', dtype=str)
```

```python
# Encode class values as integers
encoder = LabelEncoder()
encoder.fit(y_train_text)
y_train = encoder.transform(y_train_text)
y_test = encoder.transform(y_test_text)
```

```python
# One hot encode class values
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
# Reshape the data for input into CNN
num_rows = X_train.shape[0]
num_features = X_train.shape[1]
X_train = np.reshape(X_train, (num_rows, num_features, 1))
X_test = np.reshape(X_test, (X_test.shape[0], num_features, 1))
```

```python
# Define the CNN model
model = Sequential()
model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(num_features, 1)))
model.add(Conv1D(filters=64, kernel_size=3, activation='relu'))
model.add(Dropout(0.5))
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(100, activation='relu'))
model.add(Dense(y_train.shape[1], activation='softmax'))
```

```python
# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=64, verbose=1)
```

```python
# Evaluate the model on test data
score = model.evaluate(X_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
